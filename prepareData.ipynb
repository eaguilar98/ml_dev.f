```python
_toExport=[    
    'clee', 
    'id', 
    'nombre',     
    'clase_actividad',    
    'calle',     
    'ciudad',
    'municipio', 
    'entidad', 
    'colonia', 
    'cp', 
    'tipo', 
    'longitud', 
    'latitud',       
    'jiro', 
    'explicativo', 
    'trabajadores'
]
```

```python
data_df=inegi_df[_toExport].copy()
```

```python
data_df.shape
```

```python
data_df=data_df[data_df['cp']>1]
```

```python
data_df.shape
```

```python
data_df.head()
```

```python
data_df['cp']=data_df['cp'].astype(int)
```

Gouping Data

```python
data_df.isna().sum()
```

```python
#data_df.groupby(['cp','colonia','ciudad','municipio','entidad','jiro','explicativo'])['trabajadores'].sum().sort_values(ascending=False).to_csv('agrupado.csv')
```

```python
#los tacos
data_df['tacos_tortas']=np.where(
    data_df['clase_actividad'].str.contains('tacos'),1,0
)

#la comida corrida
data_df['comida_corrida']=np.where(
    data_df['clase_actividad'].str.contains('comida corrida'),1,0
)

#Los de sistemas
data_df['de_sistemas']= np.where(
        (data_df['clase_actividad'].str.contains('cómputo','computadoras')) 
        & 
        (data_df['jiro']=='servicios')
        ,1,0
        )

#la food
data_df['comida']=np.where(
    data_df['clase_actividad'].str.contains('restaurantes'),1,0
)
```

```python
losJiros=data_df['jiro'].unique()
```

```python
data_df[ (data_df['clase_actividad'].str.contains('admin')) & (~data_df['jiro'].str.contains('restaurantes|cafeterias|bufetes'))].sample(3)
```

```python
#corporativos
data_df[data_df['clase_actividad'].str.contains('corporativos')]
```

```python
#los godinez
data_df['godinez']=np.where(
    ~data_df['clase_actividad'].str.contains('restaurantes|bufetes|cafeterias|bares'),1,0
)
```

```python
data_df[data_df['godinez']==1].sample(5)
```

```python
#los contadores
data_df['contadores']=np.where(
    data_df['clase_actividad'].str.contains('contabilidad'),1,0
)
```

```python
for jiros in losJiros:    
    data_df[jiros]=np.where(
        data_df['jiro']==jiros,1,0
    )
```

```python
data_df.sample(4)
```

```python
data_df.to_csv('inegi_data_jiros_raw.csv')
```
